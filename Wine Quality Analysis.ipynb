# Preprocessing the dataset

```python
Wine_data.isnull().sum()
```

```python
# fill the missing value 

for col, value in Wine_data.items():
    Wine_data[col] = Wine_data[col].fillna(Wine_data[col].mean())
```

# Visualize the Data


```python
# use the boxplot

fig, ax = plt.subplots(ncols =6, nrows = 2, figsize=(15,10))
index  = 0
ax = ax.flatten()


for col , value in Wine_data.items():
    sns.boxplot(data=Wine_data, y=col, ax=ax[index])
    index += 1

plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)

```

```python
# Create the disc plot

fig, ax = plt.subplots(ncols =6, nrows = 2, figsize=(15,10))
index  = 0
ax = ax.flatten()


for col , value in Wine_data.items():
    sns.distplot(value, ax=ax[index])
    index += 1

plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)
```

```python
# if wine quality is 7 or above then will consider as good quality wine

Wine_data['good quality'] = [1 if x>=7 else 0 for x in Wine_data['quality']]
```

```python
# Separating the dependent and independent variable

X = Wine_data.drop(['quality','good quality'],axis=1)
y = Wine_data['good quality']
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data into train and test datasets

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.30, random_state=42)
```

```python
#Use logistic Regression model 

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_X_train = scaler.fit_transform(X_train)
```

```python
scaler_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
log_model.coef_
```

```python
y_pred = log_model.predict_proba(scaler_X_test)
```

```python
#y_pred
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_pred,y_test))
```
